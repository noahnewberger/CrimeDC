```python
# 1 - convenient function to delay between geocoding calls
geocode = RateLimiter(locator.geocode, min_delay_seconds=1.1)
```

```python
# 2- - create location column
df_mj['location'] = df['Arrest Location Block Address'].apply(geocode)
df_mj
```

```python
# 3 - create longitude, laatitude and altitude from location column (returns tuple)
df_mj['point'] = df_mj['location'].apply(lambda loc: tuple(loc.point) if loc else None)
```

```python
# 4 - split point column into latitude, longitude and altitude columns
df_mj[['latitude', 'longitude', 'altitude']] = pd.DataFrame(df_mj['point'].tolist(), index=df_mj.index)
```
